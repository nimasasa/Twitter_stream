```python
import sys
import string
import oauth2 as oauth
import urllib2 as urllib
import json


# See assignment1.html instructions or README for how to get these credentials

api_key = CONSUMER_KEY
api_secret = CONSUMER_SECRET
access_token_key = ACCESS_KEY
access_token_secret = ACCESS_SECRET

_debug = 0

oauth_token    = oauth.Token(key=access_token_key, secret=access_token_secret)
oauth_consumer = oauth.Consumer(key=api_key, secret=api_secret)

signature_method_hmac_sha1 = oauth.SignatureMethod_HMAC_SHA1()

http_method = "GET"


http_handler  = urllib.HTTPHandler(debuglevel=_debug)
https_handler = urllib.HTTPSHandler(debuglevel=_debug)

'''
Construct, sign, and open a twitter request
using the hard-coded credentials above.
'''
def twitterreq(url, method, parameters):
  req = oauth.Request.from_consumer_and_token(oauth_consumer,
                                             token=oauth_token,
                                             http_method=http_method,
                                             http_url=url, 
                                             parameters=parameters)

  req.sign_request(signature_method_hmac_sha1, oauth_consumer, oauth_token)

  headers = req.to_header()

  if http_method == "POST":
    encoded_post_data = req.to_postdata()
  else:
    encoded_post_data = None
    url = req.to_url()

  opener = urllib.OpenerDirector()
  opener.add_handler(http_handler)
  opener.add_handler(https_handler)

  response = opener.open(url, encoded_post_data)

  return response

def fetchsamples(query):
  url =  "https://api.twitter.com/1.1/search/tweets.json?q="+query

  parameters = []
  response = twitterreq(url, "GET", parameters)
  count = 0
  for line in response:
    
    print line.strip()
    count += 1
    if count>10000:
      break

if __name__ == '__main__':
  fetchsamples('microsoft')




# Test for Problem 2:
# tweet_lengths = [len(tweet) for tweet in search("microsoft")]
# print sum(tweet_lengths) // len(tweet_lengths)

# Problem 3:
# def read_scores(filename):
#    """Given a string representing a filename, open the file and return a dictionary mapping words to scores.  The file can be assumed to contain one word and one integer score on each line."""


# Test for Problem 3:
# scores = read_scores("AFINN-111.txt")
# scores["foreclosure"] + scores["masterpiece"] + scores["questionable"]
#

# Problem 4:
# def sentiment(message):
#   """Compute the sentiment of the message by adding up the scores for each word.""" 

# Test for Problem 4
# print sentiment("This significant announcement was short-sighted and disappointing, but ultimately useful.")

# Problem 5:
# def total_sentiment(list_of_strings):
#    """Return the total sentiment of a list of strings"""
#
 
# Test for Problem 5:
# print total_sentiment(["excellent fun", "terrible disaster", "foolish mistake"])


# Problem 6: No new function; modify your search function to return more pages.
# 
# Test for Problem 6:
# print len(search("university of washington"))

# Problem 7: 
# def is_from_california(tweet_object):
#   """Determine if the tweet is from California, return True or False""" 

# Test for Problem 7:
# tweetfile = open("saturday_afternoon.json", "r")
# tweets = [json.loads(line) for line in tweetfile]
# calitweets = [tweet for tweet in tweets if is_from_california(tweet)]


# Problem 8:
# def is_from_newyork(tweet_object):
#   """Determine if the tweet is from New York, return True or False""" 

# Test for Problem 8: 
# nytweets = [tweet for tweet in tweets if is_from_newyork(tweet)]


```
